### Regression algorithm

Implement Non-parametric locally weighted Regression algorithm in order to fit data points

Regression algorithm is supervised machine learning algorithm in which input x is computed with output y by mathmatical relationship such as linear relationship or quadratic relationship.Such relation where output y is linearly dependent on input x known as linear regression.

Regression algorithm is very useful algorithm of machine learning use in salary prediction,stack price prediction,bank loan prediction,exam mark prediction.Also logistic regression use to classify categorical data as classifier.

Dataset for practical :- Height and weight of people

```python
import pandas as pd,numpy as np,math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("data.csv")
```

```python
data.columns
```

```python
x=data['Height']
y=data['Weight']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
x_train=pd.DataFrame(x_train)
x_train=list(x_train.Height)
y_train=pd.DataFrame(y_train)
y_train=list(y_train.Weight)
```

```python
x_train=np.reshape(x_train,(len(x_train),1))
```

```python
model.fit(x_train,y_train)
```

```python
x_test=pd.DataFrame(x_test)
x_test=list(x_test.Height)
x_test=np.reshape(x_test,(len(x_test),1))
```

```python
y_pred=model.predict(x_test)
```
